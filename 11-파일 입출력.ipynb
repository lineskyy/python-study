# 파일 입출력
- open  함수  
    - 파일을 읽고, 열수 있는 함수
    - 파일 객체를 반환
    - `w` : 쓰기 모드
    - `a` : 추가 모드
    - `r` : 읽기 모드

- 쓰기 모드

```python
f = open("memo.txt","w") # 해당 파일이 없을경우 새로 생성
print(f)
f.close() # 파일을 열면 항상 닫아줘야한다.
```

```python
f = open("memo.txt","w")
for i in range(1,11):
    data = f"{i}번쨰 줄\n"
    f.write(data)
f.close()
```

- 쓰기 모드로 여는 순산 안에 파일 내용이 사라진다.

- 추가모드 

```python
f = open("memo.txt","a")
data = "11번째 중 \n"
f.write(data)
f.close()
```

```python
f = open("memo.txt","a")
data = "12번째 줄\n"
f.write(data)
f.close()
```

- 읽기모드

```python
f = open("memo.txt","r")
data = f.read() # 파일 전체의 내용을 읽어서 반환한다.
f.close()
print(data)
```

```python
f = open("memo.txt","r")
while True:
    data = f.readline() # 한줄씩 읽는데 줄바꿈 문자도 포함된다.
    if not data:
        break
    print(data,end="")
f.close()
```

```python
f = open("memo.txt","r")
data = f.readlines()
f.close()
print(data)
```

- `with` 문
    - 자동으로 파일을 닫아준다.

```python
with open("memo.txt","r") as f: # open 함수로 인해 반환되 파일객체를 f 변수가 받아준다.
    data = f.read()
    print(data)
```

- with 문은 파일을 자동으로 닫아 주기 때문에 다음 처럼 메소드를 실행시 오류가 발생한다.

- for 문을 이용한 방법

```python
with open("memo.txt","r") as f:
    for line in f:
        print(line,end="")
```

```python
with open("memo.txt","r") as f:
    for line in f:
        print(line,end="")
```

```python
with open("memo.txt","r") as f:
    for line in f:
        line = line.strip()
        print(line)
```

```python
with open("memo.txt","r") as f:
    for i,line in enumerate(f,1):
        print(f"No.{i} {line}",end="")
```

# JSON 파일 입출력
- 자바스크립트 에서 사용되는 객체 표현식
- 데이터를 저장하거나 전송할때 많이 사용
- 서버와 클라이언트간의 데이터 전송과정에서 많이 사용
- 파이썬의 딕셔너리와 매우 유사

```python
import json
```

```python
DATA_PATH
```

```python
with open(f"{DATA_PATH}info.json", "r") as f:
    data = json.load(f) # json 파일을 딕셔너리 형태로 반환해서 변환
type(data)
```

```python
data["관수형"] = {"age":42}
data
```

```python
with open("info.json","w") as f:
    # 한글포함시 : ensure_ascii=False
    # 들여쓰기 : indent=
    json.dump(data,f,ensure_ascii=False,indent=4)
```

# YAML (Yet Another Markup Language)
- 사람이 쉽게 익을수 있는 데이터 직렬화 양식
- 데이터를 저장하거나 전송할때 많이 사용

```python
import yaml
```

```python
data
```

```python
with open("data.yml","w") as f:
    yaml.dump(data, f, default_flow_style=False,indent = 4 , allow_unicode=True)
```

```python
with open("data.yml","r") as f:
    data1 = yaml.load(f)
type(data1)
```

```python
data1["리스트"] = [True,False,"문자열",3.14]
```

```python
with open("data1.yml","w") as f:
    yaml.dump(data1,f,default_flow_style=False,indent=4,allow_unicode=True)
```

```python
with open("data1.yml","r") as f:
    data1 = yaml.load(f)
data1
```

# pickle 모둘
- 우리가 사용하는 다양한 객체들을 파일로 저장하고 싶다면?
- 객체의 구조를 직렬화 해서 파일로 저장한다.

```python
import pickle
```

```python
data = [1,2,3,4,[10.20],"hi"]
```

```python
with open("data.pkl","wb") as f:
    pickle.dump(data,f)
```

```python
with open("data.pkl","rb") as f:
    rdata = pickle.load(f)
rdata
```

- 클래스와 인스턴스도 저장되는지

```python
class MyClass:
    def __init__(self):
        self.my_name="MyClass"

with open("myclass.pkl","wb") as f:
    pickle.dump(MyClass,f)

with open("myclass.pkl","rb") as f:
    class_ = pickle.load(f)
class_
```
