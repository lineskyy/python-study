# iterable @ iterator
- itrerable 객체: 박복 가능한 객체(list,dict,set,tuple,str,range..)
- iterator 객체: 값을 차례대로 꺼낼 수 있는 객체
    - 다 꺼내면 사라진다.
    - 꺼낼때 데이터를 메모리에 생성해주는 방식

- iter 함수
    - iterable 객체를 iterator 객체로 변환한다. 

```python
lst = [1,2,3]
iterator = iter(lst)
type(iterator)
```

- for문은 iterable,iterator 객체 모두 반복 해준다.

```python
for i in lst:
    print(i)
```

```python
for i in iterator: # iterator 객체는 다꺼내면 사라진다.
    print(i)
```

## 함수를 이용한 제너레이터 객체 만들기
- yield 문을 이용하면 된다.

```python
def gen_func():
    yield 1
    yield 2
    yield 3
```

```python
gen = gen_func()
type(gen)
```

```python
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen)) # 더이상 꺼낼수 없어 에러 발생
```

```python
def gen_func(num):
    for i in range(num):
        yield i

```

```python
gen = gen_func(5)
type(gen)
```

```python
for n in gen:
    print(n)
```

```python
lst = ["사과","딸기",None,"참외",None,"수박"] # 내가함
def gen_list(lst):
    for s in gen_list.split(None,""):
        yield s
        yield s
        yield s
        yield s
        yield s

```

```python
gen = gen_list(lst)
for i in gen:
    print(i)
```

```python
lst = ["사과","딸기",None,"참외",None,"수박"] # 쌤이함
def gen_list(lst):
    for item in lst:
        if item is not None:
            yield item
gen = gen_list(lst)
for s in gen:
    print(s)

```

```python
lst = [50,70,80] # 내가함
def gen_enumerate(param):
    for i in enumerate(lst):
        print(i)

gen_enumerate(lst)
```

```python
def gen_enumerate(param): # 썜이함
    length = len(param)
    for i in range(length):
        yield i,param[i]

lst = [50,70,80]
for i,item in gen_enumerate(lst):
    print(i,item)
```
