- else 문
  - 해당 예외가 발생하지 않을 경우 수행할 코드

````python
n = 100
for i in range(10):
    try:
        result = n/i
    except ZeroDivisionError as e:
        print(e)
    else:
        result **= 2
        print(result)
````

- finally문
    - 예외발생 여부와 상관없이 무조건 수행할 코드

````python
for n in range(3):
    try:
        result = 100/n
    except ZeroDivisionError as e:
        print(e)
    else:
        print("정상적으로 연산 되었습니다.")
    finally:
        print("처리완료")
        print(result)
````

- 예외 여부 상관없이 정수형으로 변환 및 저장완료 출력

````python
lst = ["3","5","3","3.0","10"]
num_list = []
for i in lst:
    try:
        num = int(i)
        print("정수형으로 변경합니다.")
    except ValueError as e:
        print("정수형으로 변환되지 않습니다 실수형으로 변황후에 시도하겠습니다")
        num_list.append(int(float(e)))
    else:
        print("정수변경")
    finally:
            num_list = int(num)
            print("저장완료")
            num_list
````
