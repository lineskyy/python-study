# 정규식
- 정규표현식은 복잡한 문자열을 처리할때 사용하는 기법
- 파이썬만의 고유한 기법이 아니라 문자열을 처리하는 모든곳에서 사용

```python
data = "나의 전화번호는 010-0000-0000 입니다. 너의 전화 번호는 010-1111-1111 입니다."
```

- 텍스트의 규칙을 보니 띄어쓰기 기준으로 나누자
- 전화번호의 특징이 하이픈 사이에 2개가 있다.
- 전화번호의 특징이 첫3자리,두번쨰4자리,세번째4자리 형식이다.

```python
phone_list = []
cond_list = []
for item in data.split():
    item = item.split("-")

    if len(item) == 3:
        cond = item[0].isdigit() and len(item[0]) == 3
        cond_list.appen(cond)
        cond = item[0].isdigit() and len(item[1]) == 4
        cond_list.appen(cond)
        cond = item[0].isdigit() and len(item[2]) == 4
        cond_list.appen(cond)
    
    if sum(cond_list) == 3
    phone_list.append("-".join(item))
phone_list
```

# re(regular expression) 모듈
- 파이썬의 정규표현식을 지원하기 위한 모듈

- match 함수
    - pattern 맞는 문자열로 시작하는지 검사하고 매칭이 되면 match 객체를 반환하고 그렇지 않으면 None 반환한다.

```python
import re
data = "나의 전화번호는 010-0000-0000 입니다. 너의 전화 번호는 010-1111-1111 입니다."
print(re.match("나의",data))
print(re.match("전화",data))
```

- search 함수
    - pattern에 맞는 문자열을 검사하고 매칭되면 바로 match 객체를 반환하다. 매칭이 안되면 None응 반환

```python
data = "나의 전화번호는 010-0000-0000 입니다. 너의 전화 번호는 010-1111-1111 입니다."
re.search("전화",data)
```

- findall 함수
    - 정규식과 매칭되는 모든 문자열을 리스트로 반환한다.

```python
re.findall("번호",data)
```

- sub 함수
    - 정규식과 매칭되는 부분을 치환해준다.

```python
data = "나의 전화번호는 010-0000-0000 입니다. 너의 전화 번호는 010-1111-1111 입니다."
re.sub("전화번호","핸드폰번호",data)
```

## 메타문자
- 메타문자란 원래 그문자가 가진 뜻이 아닌 특별한 용도로 사용하는 문자
- `\ ^ $ . [ ]( ) * + ? { }`

- `.`: 줄바꿈 문자(`\n`)를 제외한 모든 문자와 매치

```python
data = "abc a\nc a c"
re.findall("a.c",data)
```

- `*`: 바로 앞에 문자가 0번이상 반복할 경우 매치

```python
data = "abc ac abbc a c"
re.findall("ab*c",data)
```

`+` 바로앞에 문자가 1번 이상 반복할 경우 매치

```python
data = "abc a c abbc"
re.findall("ab+c", data)
```

```python
data = "나는 말했다:" "오늘 행복하가라거...."
```

- 반복 횟수를 제한하고 싶다면?
    - `{m/n}`: 바로 앞에 문자의 반복횠수가m번에서
 n번사이
    - `{m/n}`: 바로 팡에 문자의 반복 횟수가 m번
    - `{m/n}` 바로 팡에 문자의 반복 횟수가m번 이상
    - `{m/n}`:바로 팡에 문자의 반복 횟수가n번 이하

```python
data = "abc ac abbbc abbbbc"
re.findall("ab{1,3}c",data)
```

```python
data = "abc ac abbbc abbbbc"
re.findall("ab{,3}c",data) # 3번 이하~
```

```python
data = "abc ac abbbc abbbbc"
re.findall("ab{1,}c",data) #ab+c
```

```python
data = "abc ac abbbc abbbbc"
re.findall("ab{0,}c",data) # 번 이상~
```

```python
data = "abc ac abbbc abbbbc"
re.findall("ab{1,3}c",data) # ~ 번에서 ~ 번 까지
```

```python
data = " avc ac avvvvvbc acccvbbvvcc"
re.findall("avc{3}c",data)
```

- `?`: 반복은 아니지만 앞에 문자가 있거나 없거나 를 의미 

```python
data = "abc ac abbbc abbbbc"
re.findall("ab?c",data) #ab{0,1}c
```

- `[ ]` : [ ] 사이의 문자들과 매칭이라는 의미

```python
data = 'abc ac acc acb'
re.findall("a[bc]c",data)
```

- 대괄호 안에서는 메타문자 역할을 안한다.

```python
data = "abc a+c a.c a?c"
re.findall("a[+.?]c",data)
```

```python
data = "abc a[c a]c"
re.findall("a[\[\]]c",data)
```

```python
data = "abc ac abbbc abbbbc"
re.findall("ab{1,2}c",data)
```

```python
data = "abc a*c ab=bc ab+bc"
re.findall("a[*=+]bc",data)
```

- 하이픈(`-`) 사용하기
    - `[a-z,A-Z]` : 모든 알파벳과 매치
    - `[0-9]`: 모든 숫자와 매치

```python
data = "python 3.10,5"
re.findall("[0-9a-zA-Z]",data)
```

```python
data = "python 3.10,5"
re.findall("[0-9a-zA-Z]+",data)
```

```python
data = "파이썬 버전은 3.10.5"
re.findall("[가-홓]+",data)
```

```python
data = "파이썬 버전은 3.10.5 ㅋㅋㅋ ㅜㅜ"
re.findall("[ㅏ-ㅗ가-홓ㄱ-ㅎ]+",data)
```

- `\d` : `[0-9]`동일
- `\w` : `[a-zA-Z0-9_] 동일

```python
data = "python 3_10_5"
re.findall("\w+",data)
```

```python
re.findall("\+",data)
```

- `[^]` : not 을 의미한다.

```python
data = "python 3.10.5"
re.findall("[^\d. ]+",data)
```

- `^` : 검사 하고자 하는 문자열에 정규식에 맞는 문자로 시작하는지를 검사한다.
- `$` : 검사 하고자 하는 문자열에 정규식에 맞는 문자로 시작하는지를 검사한다.

```python
phone = "p010-0000-0000"
match = re.search("^\d{3}-\d{4}-\d{4}$",phone)
```
